```python
import os.path as op
import nibabel as nb
import numpy as np
import os
```

```python
exp_dir = op.join( os.getenv('NEURO_DATA_HOME', os.getenv('HOME')), 'MRM2014', 'tmp_tmi', 'HCP_MRM2014')
```

```python
dti_dir = op.join(exp_dir, 'ReferenceDTI' )
fa_path = op.join(dti_dir, 'FA2Nifti', 'dwi_tensor_FA_mrconvert.nii')
md_path = op.join(dti_dir, 'ADC2Nifti', 'dwi_tensor_ADC_mrconvert.nii')
seg_path = op.join(exp_dir, 'test_segm.nii.gz')
```

```python
fa = nb.load(fa_path).get_data().reshape(-1)
md = nb.load(md_path).get_data().reshape(-1)
seg = nb.load(seg_path).get_data().astype(np.uint8).reshape(-1)
```

```python
from itertools import cycle
import matplotlib.mlab as mlab
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('ggplot')

colorwheel = cycle(plt.rcParams['axes.color_cycle'])
labels = ['Thalamus', 'dGM+CSF', 'Brain stem', 'Cerebellar GM', 'WM', 'Cortical GM']

sX = (fa.max() - fa.min()) / 100
sY = (md.max() - md.min()) / 100

xmin = np.percentile(fa[fa>0.0], 2)
#xmax = np.percentile(fa[fa>0.0], 98)
xmax = fa.max()

ymin = np.percentile(md[md>0.0], 25)
ymax = md.max()
#ymax = np.percentile(md[md>0.0], 98)

X = np.linspace(xmin, xmax, num=100)
Y = np.linspace(ymin, ymax, num=100)

X, Y = np.meshgrid(X, Y)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
p = []

#fig, axs = plt.subplots(3, 3, sharex=True, sharey=True, figsize=(18,19))
#axs = axs.reshape(-1)

fig = plt.figure(figsize=(18,19))
ax = fig.gca()
# ax = fig.gca(projection='3d')
# ax.view_init(50, 30)
plt.xlim([xmin, xmax])
plt.ylim([ymin, ymax])

colors = []

for i, l in enumerate(np.unique(seg)[:-1][::-1]):
    c = colorwheel.next()
    c = colorwheel.next()
    x = fa[seg==l]
    y = md[seg==l]
    ax.scatter(x, y, c=c, alpha=.2)
    colors.append(c)

for i, l in enumerate(np.unique(seg)[:-1][::-1]):
    #ax = fig.add_subplot(3, 3, i+1, projection='3d')
    #ax.set_xlim([fa.min(), fa.max()])
    #ax.set_ylim([md.min(), md.max()])
    x = fa[seg==l]
    y = md[seg==l]
        
    mu = (np.median(x[x>0.0]), np.median(y[y>0.0]))
    #mu = (np.mean(x[x>0.0]), np.mean(y[y>0.0]))
    sample = np.vstack((x, y))
    sigma = np.cov(sample)
    
    Z = multivariate_normal.pdf(pos, mean=mu, cov=sigma)
    #p.append(ax.plot_wireframe(X, Y, Z, color=c, alpha=.5, rstride=2, cstride=2, label=labels[i]))
    
    mylevels = sorted(Z.reshape(-1))[::-1][:200][::20]
    ax.contour(X, Y, Z, levels=mylevels, alpha=.9, colors=colors[i], label=labels[i], lw=5)
    p.append(plt.Rectangle((0,0),1,1,fc=colors[i]))
    
    # cset = ax.contourf(X, Y, Z, zdir='z', color=c, alpha=.3)
    # cset = ax.contourf(X, Y, Z, zdir='x', color=c, alpha=.3)
    # cset = ax.contourf(X, Y, Z, zdir='x', alpha=.3)
    
  
ax.set_xlabel('fractional anisotropy')
ax.set_ylabel('apparent diffusion coefficient')
ax.legend(p, labels[::-1])
```
