```python

import nibabel as nb
import numpy as np
import os
import seaborn as sn
import pandas as pd
from pyacwereg import viz

exp_dir = '/media/data/TMI_mirror/experiments/data'
fa_path = op.join(exp_dir, 'fa.nii.gz')
md_path = op.join(exp_dir, 'md.nii.gz')
msk_path = op.join(exp_dir, 'mask.nii.gz')
seg_path = op.join(exp_dir, 'gold_seg.nii.gz')

im = nb.load(fa_path)
fa = nb.load(fa_path).get_data().reshape(-1)
md = nb.load(md_path).get_data().reshape(-1)
seg = nb.load(seg_path).get_data().astype(np.uint8).reshape(-1)
msk = nb.load(msk_path).get_data().astype(np.uint8).reshape(-1)

mdlims = np.array((0, 2000))
# mdlims = np.array((3e-4, 8e-4))

md[md<mdlims[0]] = np.nan
md[md>mdlims[1]] = np.nan

# md*=10e6
# mdlimssc= tuple(mdlims*10e6)
mdlimsc = tuple(mdlims)

labels = ['Thalamus', 'dGM+CSF', 'Brain stem', 'WM', 'cbGM', 'GM', 'Background']

df_pieces = []
for i, l in zip(np.unique(seg), labels):
    idxs = np.where( (seg==i) & (msk>0))
    fa_f = fa[idxs]
    md_f = md[idxs]
    d = { 'fractional anisotropy (FA)': fa_f, 'apparent diffusion coefficient (ADC)': md_f, 'tissue': [l] * len(fa_f)}
    df_pieces.append(pd.DataFrame(d))
    
    newfa = np.zeros_like(fa)
    newfa[idxs] = fa_f
    nb.Nifti1Image(newfa.reshape(im.get_shape()), im.get_affine(), im.get_header()).to_filename(op.join(exp_dir,'fa_filtered_%03d.nii.gz' % i))
    newmd = np.zeros_like(md)
    newmd[idxs] = md_f
    nb.Nifti1Image(newmd.reshape(im.get_shape()), im.get_affine(), im.get_header()).to_filename(op.join(exp_dir,'md_filtered_%03d.nii.gz' % i))

df = pd.concat(df_pieces)
df = df[df.tissue != 'Background']
```

```python
from scipy import stats as sst
import scipy  
import scikits.bootstrap as bootstrap


results = {}

for res in ['hi', 'lo']:
    results[res] = {}
    for t in ['gyrus', 'ball', 'L', 'box']:
        results[res][t] = {}
        ndf = df[df.model_type == t]
        ndf = ndf[ndf.resolution == res]
        sig = (.001 / len(ndf.surfdist_avg))

        normtest = sst.shapiro(ndf.surfdist_avg)[1]
        results[res][t]['normality'] = 'p=%g' % normtest

        resval = 2.0 if res == 'lo' else 1.0
        rstest = sst.wilcoxon(ndf.surfdist_avg - resval)[1]
        results[res][t]['accuracy'] = 'p=%g' % rstest
        results[res][t]['cohens_d'] = ndf.surfdist_avg.mean() / ndf.surfdist_avg.std()
        results[res][t]['CI0'], results[res][t]['CI1'] = bootstrap.ci(data=ndf.surfdist_avg, statfunction=scipy.median)

        
resdf_hi = pd.DataFrame(results['hi'])
resdf_low = pd.DataFrame(results['lo'])
resdf_low
```
